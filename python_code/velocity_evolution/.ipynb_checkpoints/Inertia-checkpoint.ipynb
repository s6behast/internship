```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 12, 10
import numpy as np 
import glob 
import os 


data_dir='/users/bhastings/internship/data/MW/'
file_ext= '.rot1'
```

```python
data_files=sorted(sorted(glob.glob(data_dir +'*'+ file_ext)), key=len) 

print 'There are ', len(data_files), file_ext, ' files', '\nin directory ', data_dir



def read_file_rot(f):
    col_names= ["1:t[yr]", "2:Porb[days]", "3:vsurf[km/s]", "4:Pspin[days]",
                "5:L[gcm^2/s]", "6:L-Lsync", "7:vsurf???", "8:period???"]
    #col6 is deviation in angular momentum from solid body rotation 
    
    
    
    #some files have scientific notation with D instead of E. Eg 1E32 is written as 1D32
    #code below fixes this with converters in read_csv
    
    #def D_to_E(s):
    #    if s.isdigit():
    #        return s #np.float(s.replace('D', 'E'))

    #    else:
    #        return s.replace('D', 'E')

    D_to_E = lambda x : x.replace("D", "E")
    
    converter={0:D_to_E,1:D_to_E,2:D_to_E,3:D_to_E,4:D_to_E,5:D_to_E,
                               6:D_to_E,7:D_to_E,8:D_to_E}

    
    df= pd.read_csv(f, delim_whitespace=True,comment='#', names=col_names, converters=converter)
    
    
    
    df = df.apply(pd.to_numeric, errors='coerce', axis=1) # convert all values to floats         
    df['9:Inertia[gcm^2]']= 0.5* df['5:L[gcm^2/s]'] * df['4:Pspin[days]']   /np.pi
        
    model_name= f.split('/')[-1].rsplit('.',1)[0]        
    return df, model_name


u='f9-50.mw'
df,n= read_file(data_dir+u + file_ext)     
print df.columns.tolist()

df.head()
```
