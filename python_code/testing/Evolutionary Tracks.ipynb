```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 8, 6
import numpy as np 
import glob 

data_dir='/users/bhastings/internship/data/MW/'
file_ext= '.dat2'
```

## Define function to read .dat2 data ##

```python
data_files=sorted(sorted(glob.glob(data_dir +'*'+ file_ext)), key=len) 

print 'There are ', len(data_files), file_ext, ' files', '\nin directory ', data_dir



def read_file(f):
    model_name= f.split('/')[-1].rsplit('.',1)[0]
    wanted_cols=['1:t[s]', '2:M/Msun', '3:Teff[K]', '4:log(L/Lsun)','5:R/Rsun','6:log(Mdot)[Msun/yr]',
                 '8:v_crit[km/s]', '9:v_surf[km/s]', '11:H','12:He','26:H_massfr', '27:He_massfr']

    if "-0" in str(f): # if file is for NON rotating star, remove vrot and crit vel from col names 
            col_names= ['1:t[s]', '2:M/Msun', '3:Teff[K]', '4:log(L/Lsun)', '5:R/Rsun', '6:log(Mdot)[Msun/yr]', '7:logg[cgs]', 
                        '10:P[days]', '11:H', '12:He', '13:Li', '14:Be', '15:B', '16:C', '17:N', '18:O', 
                    '19:F', '20:Ne', '21:Na', '22:Mg', '23:Al', '24:Si', '25:Fe', '26:H_massfr', '27:He_massfr']
    else: 
        col_names= ['1:t[s]', '2:M/Msun', '3:Teff[K]', '4:log(L/Lsun)', '5:R/Rsun', '6:log(Mdot)[Msun/yr]', '7:logg[cgs]', '8:v_crit[km/s]', 
                '9:v_surf[km/s]', '10:P[days]', '11:H', '12:He', '13:Li', '14:Be', '15:B', '16:C', '17:N', '18:O', 
                '19:F', '20:Ne', '21:Na', '22:Mg', '23:Al', '24:Si', '25:Fe', '26:H_massfr', '27:He_massfr']

    df= pd.read_csv(f, delim_whitespace=True,comment='#', names=col_names, usecols=wanted_cols)
    return df, model_name




#select a file to load to test process  
f=data_files[30]
print f
df,model_name= read_file(f)
df.head()
```

## plot evolutionary tracks for models of a particular mass ## 

```python
mass_list=[3,5,7, 9, 12, 15, 20, 25, 30,40, 50, 60, 80, 100 ] # list of model masses 

target_mass=12


plt.clf()
for m in mass_list[0:1]:
    plt.clf()
    target_mass=m
    
    for f in data_files:
        if f.find('f'+str(target_mass)+'-') !=-1: #if file name contains target mass
            df,mod_nam=read_file(f)
            #print f
            x,y= [np.log10(df['3:Teff[K]']), df['4:log(L/Lsun)']]
            
            if '9:v_surf[km/s]' in df.columns:
                crit_frac= '{0:.2f}'.format([df['9:v_surf[km/s]']/df['8:v_crit[km/s]']][0][0]) # critical velocity of 1st evolutionary model
                
                
            else:
                crit_frac=0.0
                
            plt.plot(x,y, '-',label=crit_frac)
        

    plt.title("Evolutionary Tracks for "+str( target_mass) +" $M_{\odot}$  models")
    plt.xlabel(r'$Log (T_{eff}/K)$', size=16)
    plt.ylabel(r'$ Log (L/L_\odot)$',size=16)
    
    leg = plt.legend(fontsize = 'large', loc='upper right', bbox_to_anchor=(1.25, 1.0))
    leg.set_title(("Initial " +r'$\frac{V}{V_{crit}}$'), prop = {'size':18})

    #plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),ncol=5, title=r'$\frac{V}{V_{crit}}$' ,prop={'size':18})
    
    plt.gca().invert_xaxis() # invert xaxis 

    #plt.xlim(4.2,4.5) H-burinng
    #plt.xlim(3.5,3.7) He-burning 

    plt.show()
    
    #plt.savefig(("./evol_tracks/" +str(target_mass)+"evol_track.png"), dpi=300 ,bbox_inches='tight')
```

```python

#model_name= f.split('/')[-1].rsplit('.',1)[0]




def plot_HR(HR_data,model_file, save_fig, markevery):
    plt.figure(figsize=(10,8))
    
    x,y,c=[np.log10(HR_data['3:Teff[K]'])[0::markevery], 
           HR_data['4:log(L/Lsun)'][0::markevery], 
           (HR_data['9:v_surf[km/s]']/HR_data['8:v_crit[km/s]'])[0::markevery]]
    
    m=model_file.split('/')[-1].split('.')[0].split('-')[0][1:]
    v=model_file.split('/')[-1].split('.')[0].split('-')[1]

    plt.ylim(y.min()*0.98, y.max()*1.02)
    plt.xlim(x.min()*0.99, x.max()*1.02)
    plt.title("HR Diagram for "+str(m) +" $M_{\odot}$ , V= "+str(v)   +" km/s model")
    plt.xlabel(r'$Log (T_{eff}/K)$', size=16)
    plt.ylabel(r'$ Log (L/L_\odot)$',size=16)

    plt.scatter(x,y, c=c, cmap='viridis' )

    plt.gca().invert_xaxis() # invert xaxis 

    cbar = plt.colorbar()
    cbar.set_label(r'$\frac{V}{V_{crit}}$', rotation=0, size=24, labelpad=40, y=0.5)

    #p#lt.ylim(3.9,4.11)
    #plt.ylim(1.8,2.4)
    
    plt.plot()
    
    
    if save_fig:
        plt.savefig("HR_ZAMS.png", dpi=600)
    else: 
        plt.show()
        
%matplotlib inline

u='f15-250.mw.dat2'
df,n= read_file(data_dir+u)        
#plot_HR(df,n, False,30)
print df.columns
u='f15-400.mw.dat2'
plot_HR(read_file(data_dir+u)[0],read_file(data_dir+u)[1], False,30)


#plt.plot(df['1:t[s]'], (df['9:v_surf[km/s]']), label='surf')#/df['8:v_crit[km/s]']))
#plt.plot(np.log10(df['5:R/Rsun']), (np.log10(df['8:v_crit[km/s]'])), label='crit')#/df['8:v_crit[km/s]']))
#plt.plot(np.log10(df['5:R/Rsun']), (np.log10(df['9:v_surf[km/s]'])), label='surf')#/df['8:v_crit[km/s]']))
#plt.legend()
#plt.show()

#plt.plot(df['1:t[s]'], (df['5:R/Rsun']), label='rad')#/df['8:v_crit[km/s]']))
#plt.legend()
#plt.show()
```
