```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 8, 6
import numpy as np 
import glob 

data_dir='/users/bhastings/internship/data/MW/'
file_ext= '.plot1'
```

## load in a single plot file ##

```python
data_files=sorted(glob.glob(data_dir +'*'+ file_ext))

print 'There are ', len(data_files), file_ext, ' files', '\nin directory ', data_dir

#select a file to load 
f=data_files[10]
print f


#some files have scientific notation with D instead of E. Eg 1E32 is written as 1D32
#code below fixes this with converters in read_csv
def D_to_E(s):
    if s.isdigit():
        return s #np.float(s.replace('D', 'E'))
    
    else:
        return s.replace('D', 'E')

def read_file(f):
    #print "using file ", f 
    col_names= ["t[yr]","Tc[1e8K]","Yc","L_H[log(L_H/Lsun)]","L_He[log(L_He/Lsun)]","M[Msun]","??",
                "Teff[K]","L[log(L/Lsun)]", "log(rho_c)[log(gcm-3)]", "L_C[log(L_C/Lsun)]", "L_Nu[log(L_Nu/Lsun)]", 
                "Mdot_wind[Msun yr-1]", "T_max[K]", "rho@T_max[gcm-3]","M_r@T_max[Msun]"  ]
    dat= pd.read_csv(f, delim_whitespace=True,comment='#', names=col_names,
                    converters={0:D_to_E,1:D_to_E,2:D_to_E,3:D_to_E,4:D_to_E,5:D_to_E,
                               6:D_to_E,7:D_to_E,8:D_to_E,9:D_to_E,10:D_to_E,11:D_to_E,
                               12:D_to_E,13:D_to_E,14:D_to_E,15:D_to_E})


    
    dat = dat.apply(pd.to_numeric, errors='coerce', axis=1) # convert all values to floats 
    return dat 

dat=read_file(f)

dat.head()
```

## find all MS models of the plot file ##

```python

def MS_finder(dat, plot):
    i=200 # number of data points to truncate to 
    tol= 2e-4 # boundary for finding L_H/L =1 
    #print "condition to be on MS is L_H/L to be within ", 1-tol, 1+tol

    #divide h burning luminosity by total luminosity for every model 
    dat['H_lum_frac']=dat['L_H[log(L_H/Lsun)]']/dat['L[log(L/Lsun)]']
    x,y= [dat['t[yr]'][0:i], dat['H_lum_frac'][0:i]]

    if plot==True:
        #plot results 
        plt.plot(x,y, 'ko-')
        plt.axhline(1+tol, color='r')
        plt.axhline(1-tol, color='r')
        #plot formatting 
        plt.ylim(0.995,1.005)
        plt.xlabel('t/yr')
        plt.ylabel('L_H/L')
        plt.show()


    #find models with age greater than 0 and with h burning fractions within tolerance tol


    MS_models=dat[(dat['t[yr]'])>0][dat['H_lum_frac'].between(1-tol, 1+tol)]

    MS_models.head()
    
    return MS_models


MS_models= MS_finder(dat, True)
MS_models.head()
```

## find ZAMS model ## 
sometimes the first MS model selected just has by chance a H-burning luminosity equal to the total luminosity (eg f20-400.mw.plot1). 

To check for true MS, the following models must be checked for the MS condition. 

```python
def ZAMS_finder(df):

    # loop over indexes in MS models dataframe. 
    #When a row is found that has max_cond_count successive MS rows (in the original dataframe), this row is taken as ZAMS


    index_counter=0
    condition_counter=0
    max_cond_count=20


    for index in df.index.values.tolist():
        next_index= df.index.values.tolist()[index_counter+1]
        #print index, next_index
        if next_index - index ==1 : 
            #print "poss zams"
            condition_counter=condition_counter+1

            if condition_counter >=max_cond_count: 
                #print "ZAMS found @ index ",  df.index.values.tolist()[index_counter- max_cond_count +1]
                ZAMS_index= df.index.values.tolist()[index_counter- max_cond_count +1]
                break
        index_counter=index_counter+1

    ZAMS_model= df.loc[ZAMS_index]
    
    return ZAMS_model


ZAMS_model= ZAMS_finder(MS_models)
ZAMS_model.head()

print ZAMS_model['t[yr]']
```

## Now iterate process to run over all files##

```python
data_files=sorted(glob.glob(data_dir +'*'+ file_ext))

ZAMS_ages=pd.DataFrame(columns=["file_name","M[Msun]", "Vrot[km/s]", "ZAMS_age[yr]"])
#########################################################
#REMEMBER TO REMOVE LIMIT FROM 0 TO 10!!!!!!!!!!!!!!!!!!!!
for f in data_files:
    
    #load data 
    df=read_file(f)
    #print f
    #print df.head()
    #find MS models 
    MS=MS_finder(df, False)
    #find ZAMS model
    ZAMS=ZAMS_finder(MS)
    
    #add ZAMS model to dataframe (or maybe just the index and time(REMEMBER TO CONVERT TO SECONDS!!)?? )
    
    m=f.split('/')[-1].split('.')[0].split('-')[0][1:]
    v=f.split('/')[-1].split('.')[0].split('-')[1]
    file_name= f.split('/')[-1].rsplit('.',1)[0]
    #print m,v, ZAMS['t[yr]']
    single_ZAMS= pd.DataFrame([[file_name,m,v,ZAMS['t[yr]']]],columns=["file_name","M[Msun]", "Vrot[km/s]", "ZAMS_age[yr]"] )
    
    ZAMS_ages=ZAMS_ages.append(single_ZAMS)


    
ZAMS_ages["ZAMS_age[s]"]= ZAMS_ages["ZAMS_age[yr]"] * (60**2 * 24 * 365) # convert s to years 
ZAMS_ages.head()
```

```python
import sys


def progress(count, total, status=''):
    bar_len = 60
    filled_len = int(round(bar_len * count / float(total)))

    percents = round(100.0 * count / float(total), 1)
    bar = '=' * filled_len + '-' * (bar_len - filled_len)

    sys.stdout.write('[%s] %s%s ...%s\r' % (bar, percents, '%', status))
sys.stdout.flush() # As suggested by Rom Ruben (see: http://stackoverflow.com/questions/3173320/text-progress-bar-in-the-cons
```
